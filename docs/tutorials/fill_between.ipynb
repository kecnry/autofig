# Autofig fill_between

```python
import autofig
import numpy as np
```

```python
autofig.inline()
```

Let's generate some fake data

```python
x = np.linspace(0, 10, 101)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
autofig.fill_between(x, np.array([y1, y2]).T, i='x', color='blue', uncover=True, trail=0.1)
```

```python
fig = autofig.draw(show=True)
```

```python
autofig.draw(show=True, i=6.01)
```
